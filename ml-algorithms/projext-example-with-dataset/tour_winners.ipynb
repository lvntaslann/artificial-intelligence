```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder
```

```python
df = pd.read_csv('Tour_Winners_data_1.csv')
```

```python
df.head()
```

```python
df.info()
```

```python
df = df.dropna(axis=1)
df.isnull().sum()
```

```python
df['Team'].unique()
```

```python
df.info()
```

```python
labelEncoder = LabelEncoder()
df['Team']=labelEncoder.fit_transform(df['Team'])
df['Winner']=labelEncoder.fit_transform(df['Winner'])
df['Country']=labelEncoder.fit_transform(df['Country'])
df['Tour_overall_length_(km)']=labelEncoder.fit_transform(df['Tour_overall_length_(km)'])
df['rider_type_(PPS)']=labelEncoder.fit_transform(df['rider_type_(PPS)'])
df.head()
```

```python
# tüm verileri sayısal formata çevirildi
df.info()
```

```python
df.isnull().sum()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
```

```python
y = df['Winner']
x = df.drop(['Winner'],axis=1)
print(x.shape)
print(y.shape)
```

```python
from sklearn.preprocessing import StandardScaler,MinMaxScaler
standard_scaler = StandardScaler()
min_max_scaler = MinMaxScaler()

x_standardScaler = standard_scaler.fit_transform(x)
x_min_maxscaler = min_max_scaler.fit_transform(x)
x_standardScaler = pd.DataFrame(x_standardScaler)
x_min_maxscaler = pd.DataFrame(x_min_maxscaler)
print(x_standardScaler.head())
print(x_min_maxscaler.head())
```

```python
x_standardScaler_train,x_standardScaler_test,y_train,y_test = train_test_split(x_standardScaler,y,train_size = 0.85,random_state = 6)
```

```python
print(x_standardScaler_train.shape)
print(x_standardScaler_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
model = RandomForestClassifier()
model.fit(x_standardScaler_train,y_train)
```

```python
model.score(x_standardScaler_test,y_test)
```

```python
x_minMaxScaler_train,x_minMaxScaler_test,y_train,y_test = train_test_split(x_min_maxscaler,y,train_size = 0.85,random_state = 6)
```

```python
model2 = RandomForestClassifier()
model2.fit(x_minMaxScaler_train,y_train)
```

```python
model2.score(x_minMaxScaler_test,y_test)
```

```python
from sklearn.linear_model import LogisticRegression
model3 =LogisticRegression()
model3.fit(x_minMaxScaler_train,y_train)
```

```python
model3.score(x_minMaxScaler_test,y_test)
```
