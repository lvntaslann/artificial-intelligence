```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
df['takim'] = le.fit_transform(df['takim'])
print("Etiketler:", le.classes_)
```

```python
# 0->besiktas 1->bursa 2->fenerbahce 3->galatasaray 4->trabzonspor
df.head()
```

```python
from sklearn.model_selection import train_test_split
x = df.drop(['puan','yil'],axis = 1)
y = df['puan']
x=x.astype('float32')
x_train,x_test,y_train,y_test = train_test_split(x,y,train_size = 0.80,random_state = 42)
```

```python
print(x_train.shape)
print(y_train.shape)
```

```python
from keras.models import Sequential
from keras.layers import Dense

model = Sequential()
model.add(Dense(64,activation="relu"))
model.add(Dense(32,activation="relu"))
model.add(Dense(16,activation="relu"))
model.add(Dense(1))
```

```python
model.compile(optimizer="adam",loss="mse",metrics=["mae"])
```

```python
model.fit(x_train,y_train,epochs=256,batch_size=16,validation_split=0.24)
```

```python
mse, mae = model.evaluate(x_test, y_test)
print(f"Test Kayıp (Loss): {mse}")
print(f"Test Ortalama Mutlak Hata (MAE): {mae}")
```
