```python
def fonksiyon(x,y):
    return ((x**2+y**2)/2)*np.exp(-(x**2-y**3)/4)
```

```python
x=np.arange(-1.0,1.0,0.01)
y=np.arange(-1.0,1.0,0.01)
a,b = np.meshgrid(x,y)
z=fonksiyon(a,b)
im = plt.imshow(z,cmap=plt.cm.RdBu)
seviyeler = np.arange(-1,1.5,0.2)
cizim = plt.contour(z,seviyeler,linewidths=1,cmap=plt.cm.Set2)
plt.clabel(cizim,inline=True,fmt='%1.1f',fontsize=10)
plt.colorbar(im,aspect=20,shrink=0.6,orientation='vertical')
plt.title('$z=((x^2+y^3)/2) e^{-(x^2-y^2)/4}$')
plt.show()
```

```python
from mpl_toolkits.mplot3d import Axes3D

x=np.arange(-1.0,1.0,0.01)
y=np.arange(-1.0,1.0,0.01)
a,b=np.meshgrid(x,y)
z=fonksiyon(a,b)
sekil = plt.figure()
ax=Axes3D(sekil)
yuzey = ax.plot_surface(a,b,z,rstride=1,cstride=1,cmap=plt.cm.RdBu,linewidth=0,antialiased=False)
sekil.colorbar(yuzey,shrink=0.6)
plt.title('$z=((x^2+y^3)/2) e^{-(x^2-y^2)/4}$')
```

```python
from collections import namedtuple
from pprint import pprint as pp
from math import floor

DalYaprak = namedtuple('Dal', 'veri, yaprakdigits')

veriseti = DalYaprak((46,42,38,16,35,10,42,42,26,50,60,42,26,46,10,70,46,26,34,26,
           46,38,58,46,70,54,50,12,46,50,26,62,14,50,50,54,50,50,50,54,
           42,66,42,58,30,62,46,36,22,24,54,50,40,50,40,40,28,46,72,42,
           54,18,54,42,42,46,34,42,50,54,30,46,58,48,46,36,36,66,38,34),
             1.0)

def dalyaprak(dal):
    m = []
    ic = int(10**int(dal.yaprakdigits))
    for veri in sorted(dal.veri):
        veri = int(floor(veri))
        dl, ypr = divmod(veri,ic)
        m.append( (int(dl), int(ypr)) )
    dallar, yapraklar = list(zip(*m))
    dalGenislik = max(len(str(x)) for x in dallar)
    yaprakGenislik = max(len(str(x)) for x in yapraklar)
    sonKok, cikti = min(dallar) - 1, []
    for s,l in m:
        while sonKok < s:
            sonKok += 1
            cikti.append('\n%*i |' % ( dalGenislik, sonKok))
        cikti.append(' %0*i' % (yaprakGenislik, l))
    return ''.join(cikti)

if __name__ == '__main__':
    print( dalyaprak(veriseti))
```
