```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
sirket = ["Google","Amazon","Microsoft","Facebook"]
kazanc = [90,136,89,27]
```

```python
plt.bar(sirket,kazanc)
```

```python
xpos = np.arange(len(sirket))
xpos
```

```python
xpos+0.2
```

```python
plt.bar(xpos,kazanc)
```

```python
plt.ylabel("Kazanc(milyar)")
plt.xlabel("irket")
plt.title("ABD teknoloji irketleri kazanlar")
plt.bar(sirket,kazanc,label='ciro')
plt.legend()
```

```python
kar = [40,2,34,12]
```

```python
plt.ylabel('Kazan (Milyar)')
plt.xlabel('irket')
plt.title('ABD Teknoloji irketleri Kazanlar')
plt.bar(sirket,kazanc, label = 'Ciro')
plt.bar(sirket,kar, label = 'Kar')
plt.legend()
```

```python
xpos = np.arange(len(sirket))
plt.xticks(xpos,sirket)

plt.ylabel('kazanc(milyar)')
plt.xlabel('Sirket')
plt.title('ABD teknoloji irketleri kazanlar')
plt.bar(xpos-0.2,kazanc,width=0.4,label='Ciro')
plt.bar(xpos+0.2,kar,width=0.4,label='Kar')
plt.legend()
```

```python
species = ("Adelie","Chinstrap",'Gentoo')
penguin_means = {
    'Bill Depth' :(18.35,18.43,14.98),
    'Bill Length' :(38.79,48.83,47.50),
    'Flipper Length' :(189.95,195.82,217.19)
}

x = np.arange(len(species))
width = 0.25
multiplier = 0

fig, ax = plt.subplots(layout="constrained")

for attribute, measurement in penguin_means.items():
    offset = width*multiplier
    rects = ax.bar(x+offset,measurement,width,label=attribute)
    ax.bar_label(rects,padding=3)
    multiplier=+1


ax.set_ylabel('Length(mm)')
ax.set_title('penguin attributes by species')
ax.set_xticks(x+width,species)
ax.legend(loc="upper left", ncols=3)
ax.set_ylim(0,250)

plt.show()
```

## Pie

```python
harcamalar = [1400,600,300,410,250]
harcama_kalemleri = ['kira','yemek','tel/internet','arac/yakit','dier']
```

```python
plt.pie(harcamalar)
plt.show()
```

```python
plt.pie(harcamalar,labels=harcama_kalemleri)
plt.show()
```

```python
plt.axis('equal')
plt.pie(harcamalar,labels=harcama_kalemleri,radius=1.5)
plt.show()
```

```python
plt.axis('equal')
plt.pie(harcamalar, labels=harcama_kalemleri, radius=1.5, autopct='%0.4f%%')
plt.show()
```

```python
plt.axis('equal')
plt.pie(harcamalar, labels=harcama_kalemleri, radius=1.5, autopct='%0.2f%%', shadow=True)
plt.show()
```

```python
plt.axis('equal')
plt.pie(harcamalar, 
        labels=harcama_kalemleri, 
        radius=1.5, 
        autopct='%0.2f%%', 
        shadow=True,
        explode=[0,0.2,0,0.2,0]
       )
plt.show()
```

```python
plt.axis('equal')
plt.pie(harcamalar, 
        labels=harcama_kalemleri, 
        radius=1.5, 
        autopct='%0.2f%%', 
        shadow=True,
        explode=[0,0,0.5,0,0],
        startangle=180
       )
plt.show()
```

```python
plt.axis('equal')
plt.pie(harcamalar, 
        labels=harcama_kalemleri, 
        radius=1.5, 
        autopct='%0.2f%%', 
        shadow=True,
        explode=[0,0,0.5,0,0],
        startangle=0
       )
plt.savefig('piechart1.png')
```

```python
plt.axis('equal')
plt.pie(harcamalar, 
        labels=harcama_kalemleri, 
        radius=1.5, 
        autopct='%0.2f%%', 
        shadow=True,
        explode=[0,0,0,0,0],
        startangle=0
       )
plt.savefig('piechart3.png', bbox_inches='tight')
```
