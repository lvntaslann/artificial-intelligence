```python
import pandas as pd
```

```python
df = pd.read_csv('audi.csv')
```

```python
df.head()
```

```python
df = df.drop(['index','href','MileageRank','PriceRank','PPYRank','Score'],axis = 1)
df.head()
```

```python
df.columns=["yil","kasa","mil","motor","ps","vites","yakit","sahip","fiyat","ppy"]
```

```python
df.head()
```

```python
df['motor'] = df.motor.str.replace('L','')
```

```python
df['motor'] = pd.to_numeric(df.motor)
df.motor
```

```python
df=pd.get_dummies(df,columns=['kasa','vites','yakit'],drop_first=True)
```

```python
df.head()
```

```python
df.info()
```

```python
df.head()
```

```python
df['vites_Manual']
```

```python
df['vites_Manual'] = df.vites_Manual.astype(int)
df['yakit_Petrol'] = df.yakit_Petrol.astype(int)
df.head()
```

```python
X = df.drop(['fiyat'],axis = 1)
y = df['fiyat']
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,train_size = 0.70,random_state = 22)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
model = LinearRegression()
model.fit(X_train,y_train)
```

```python
model.score(X_test,y_test)
```

```python
model.predict([[2017,30000,1.6,110,1,2600,0,1]])
```
