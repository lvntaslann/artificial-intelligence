```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv('vize.csv')
df.head()
```

```python
df = df.loc[:, ~df.columns.str.contains('Unnamed')]
df.head()
```

```python
for i in range(len(df.columns)):
    column = df.columns[i]
    print(column)
    print(df[column].unique())
```

```python
import re
def temizle(liste):
    string = ''.join(liste)
    string = re.sub(r'\([^)]*\)', '', string)
    return string.strip()

df = df.applymap(temizle)
df.head()
```

```python
print(df['Hususi Pasaport'].value_counts())
print(df['Umuma Mahsus Pasaport'].value_counts())
print(df['Diplomatik Pasaport'].value_counts())
print(df['Hizmet Pasaportu'].value_counts())
```

```python
df = df[
    (df['Hususi Pasaport'] != 'Schengen vizesi gerekli') &
    (df['Umuma Mahsus Pasaport'] != 'Schengen vizesi gerekli') & 
    (df['Umuma Mahsus Pasaport']!='Sınırda vize') &
    (df['Diplomatik Pasaport'] != 'Schengen vizesi gerekli') &
    (df['Hizmet Pasaportu'] != 'Schengen vizesi gerekli')
]

print(df['Hususi Pasaport'].value_counts())
print(df['Umuma Mahsus Pasaport'].value_counts())
print(df['Diplomatik Pasaport'].value_counts())
print(df['Hizmet Pasaportu'].value_counts())
```

```python
df.isnull().sum()
```

```python
if df.duplicated == True:
   print('Tekrar eden veri var')
else:
    print('Tekrar eden veri yok')
```

```python
df.info()
```

```python
from sklearn.preprocessing import LabelEncoder

ulke_lb = LabelEncoder()
umuma_lb = LabelEncoder()
diplomatik_lb = LabelEncoder()
hususi_lb = LabelEncoder()
hizmet_lb = LabelEncoder()

df['Ülke'] = ulke_lb.fit_transform(df['Ülke'])
df['Hususi Pasaport'] = hususi_lb.fit_transform(df['Hususi Pasaport'])
df['Umuma Mahsus Pasaport'] = umuma_lb.fit_transform(df['Umuma Mahsus Pasaport'])
df['Diplomatik Pasaport'] = diplomatik_lb.fit_transform(df['Diplomatik Pasaport'])
df['Hizmet Pasaportu'] = hizmet_lb.fit_transform(df['Hizmet Pasaportu'])
```

```python
df.head()
```

```python
print("Ülke Sayısal Değerler:", df['Ülke'].tolist())
print("Ülke Orijinal Değerler:", ulke_lb.inverse_transform(df['Ülke']))

print("\nHususi Pasaport Sayısal Değerler:", df['Hususi Pasaport'].tolist())
print("Hususi Pasaport Orijinal Değerler:", hususi_lb.inverse_transform(df['Hususi Pasaport']))

print("\nUmuma Mahsus Pasaport Sayısal Değerler:", df['Umuma Mahsus Pasaport'].tolist())
print("Umuma Mahsus Pasaport Orijinal Değerler:", umuma_lb.inverse_transform(df['Umuma Mahsus Pasaport']))

print("\nDiplomatik Pasaport Sayısal Değerler:", df['Diplomatik Pasaport'].tolist())
print("Diplomatik Pasaport Orijinal Değerler:", diplomatik_lb.inverse_transform(df['Diplomatik Pasaport']))

print("\nHizmet Pasaportu Sayısal Değerler:", df['Hizmet Pasaportu'].tolist())
print("Hizmet Pasaportu Orijinal Değerler:", hizmet_lb.inverse_transform(df['Hizmet Pasaportu']))
```

```python
df.info()
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
model = RandomForestClassifier(n_estimators=100)
x = df.drop('Hizmet Pasaportu',axis=1)
y = df['Hizmet Pasaportu']
x_train,x_test,y_train,y_test = train_test_split(x,y,train_size = 0.80,random_state = 42)
model.fit(x_train,y_train)
```

```python
model.score(x_test,y_test)
```
