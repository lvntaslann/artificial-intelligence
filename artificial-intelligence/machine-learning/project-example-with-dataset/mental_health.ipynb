```python
df.rename(columns={
    'age':'yas',
    'gender':'cinsiyet',
    'employment_status':'istihdam_durumu',
    'work_environment':'calisma_ortami',
    'mental_health_history':'ruhsal_saglik_gecmisi',
    'seeks_treatment':'tedavi_arama',
    'stress_level':'stres_seviyesi',
    'sleep_hours':'uyku_saati',
    'physical_activity_days':'fiziksel_aktivite_gunleri',
    'anxiety_score':'anksiyete_puani',
    'social_support_score':'sosyal_destek_puani',
    'productivity_score':'uretkenlik_puani',
    'mental_health_risk':'ruhsal_saglik_riski',
    },inplace=True)

df.head()
```

```python
df.info()
```

```python
data_list = ['cinsiyet','calisma_ortami','istihdam_durumu'
             ,'ruhsal_saglik_gecmisi','tedavi_arama','ruhsal_saglik_riski']
for index in range(6):
    print(f"{data_list[index]}: {df[data_list[index]].unique()}")
```

```python
from sklearn.preprocessing import LabelEncoder
labelEncoder = LabelEncoder()
encoders = {}
for index in data_list:
    le = LabelEncoder()
    df[index] = le.fit_transform(df[index])
    encoders[index] = le
```

```python
df.info()
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler

standart_scaler = StandardScaler()
x = df.drop(['ruhsal_saglik_riski'],axis=1)
y = df['ruhsal_saglik_riski']

x_standartscaler = standart_scaler.fit_transform(x)
dataframe_x = pd.DataFrame(x_standartscaler)
print(x.shape)
print(y.shape)
print(dataframe_x)
x_train,x_test,y_train,y_test = train_test_split(x,y,train_size=0.70,random_state=42)
print(x_train.shape)
print(y_train.shape)
print(x_test.shape)
print(y_test.shape)
```

```python
model = RandomForestClassifier()
model.fit(x_train,y_train)
```

```python
model.score(x_test,y_test)
```

```python
from sklearn.linear_model import LogisticRegression
model2 = LogisticRegression()
model2.fit(x_train,y_train)
```

```python
model2.score(x_test,y_test)
```
