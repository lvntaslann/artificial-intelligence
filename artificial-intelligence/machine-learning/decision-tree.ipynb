```python
import pandas as pd
import numpy as np
from sklearn.tree import DecisionTreeClassifier,export_graphviz
import graphviz
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv('heart_attack_prediction_dataset.csv')
```

```python
df.head()
```

```python
df["Sex"] = df["Sex"].replace('Male', 0)
df["Sex"] = df["Sex"].replace('Female', 1)
df
```

```python
df.info()
```

```python
df["Exercise Hours Per Week"] = df["Exercise Hours Per Week"].astype(int)
df["Sedentary Hours Per Day"] = df["Sedentary Hours Per Day"].astype(int)
df["BMI"] = df["BMI"].astype(int)
df.info()
```

```python
df['Diet']
```

```python
df['Diet'] = df['Diet'].replace('Average',0)
df['Diet'] = df['Diet'].replace('Healthy',1)
df['Diet'] = df['Diet'].replace('Unhealthy',2)
df['Diet']
```

```python
df.info()
```

```python
df.columns
```

```python
df = df.drop(['Patient ID'], axis=1)
X = df.drop(['Heart Attack Risk', 'Hemisphere', 'Continent', 'Country', 'Blood Pressure'], axis=1)
y = df['Heart Attack Risk']
```

```python
X
```

```python
X.shape
```

```python
y.shape
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,train_size =0.75,random_state = 10)
```

```python
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
model = DecisionTreeClassifier()
model.fit(X_train,y_train)
```

```python
model.score(X_test,y_test)
```

```python
from sklearn.linear_model import LogisticRegression

model2 = LogisticRegression()
model2.fit(X_train,y_train)
```

```python
model2.score(X_test,y_test)
```
