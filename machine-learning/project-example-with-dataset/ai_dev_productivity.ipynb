```python
df.rename(columns={'hours_coding':'kodlama_saati',
                   'coffe_intake_mg':'kahve_alimi_mg',
                  'distractions':'dikkat_dagiticilar',
                   'sleep_hours':'uyku_saati',
                   'bugs_reported':'bildirilen_hatalar',
                   'ai_usage_hours':'ai_kullanim_saatleri',
                   'cognitive_load':'bilissel_yukleme',
                   'task_success':'gorev_basarisi'
                  },inplace=True)

df.head()
```

```python
df.dropna()
```

```python
x = df.drop(columns=['gorev_basarisi'])
y = df['gorev_basarisi']
print(x)
print(y)
```

```python
from sklearn.preprocessing import MinMaxScaler
min_max_scaler = MinMaxScaler()
x_min_maxscaler = min_max_scaler.fit_transform(x)
df_x_min_maxscaler = pd.DataFrame(x_min_maxscaler)
print(df_x_min_maxscaler.head())
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x_min_maxscaler,
                                                 y,
                                                 train_size=0.70,
                                                 random_state=42)
```

```python
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
from sklearn.ensemble import RandomForestClassifier
model = RandomForestClassifier()
model.fit(x_train,y_train)
```

```python
model.score(x_test,y_test)
```

```python
from sklearn.linear_model import LogisticRegression
model2 = LogisticRegression()
model2.fit(x_train,y_train)
```

```python
model2.score(x_test,y_test)
```
