```python
df = df.dropna()
```

```python
df.head()
```

```python
df.shape
```

```python
df.columns = ['seyahat_mesafesi','gunun_saati','haftanin_gunu','yolcu_sayisi','trafik_kosullari','hava_durumu','temel_ucret','km_basina_ucret','dk_basina_ucret','seyahat_suresi','seyahat_ucreti']
df.head()
```

```python
df.gunun_saati
```

```python
gunun_saati = df.gunun_saati.unique()
print(gunun_saati)
```

```python
df['gunun_saati'] = df['gunun_saati'].replace({'Morning':0,'Evening':1,'Afternoon':2,'Night':3})
df.head(3)
```

```python
haftanin_gunu = df.haftanin_gunu.unique()
print(haftanin_gunu)
```

```python
df['haftanin_gunu'] = df['haftanin_gunu'].replace({'Weekday':0,'Weekend':1})
df.head(3)
```

```python
hava_durumu = df.hava_durumu.unique()
print(hava_durumu)
```

```python
df['hava_durumu'] = df['hava_durumu'].replace({'Clear':0,'Rain':1,'Snow':2})
df.head(3)
```

```python
trafik_kosullari = df.trafik_kosullari.unique()
print(trafik_kosullari)
```

```python
df['trafik_kosullari'] = df['trafik_kosullari'].replace({'Low':0,'High':1,'Medium':2})
df.head(3)
```

```python
df.info()
```

```python
x = df.drop(['seyahat_ucreti'],axis = 1)
y = df['seyahat_ucreti']
```

```python
print(x.shape)
print(y.shape)
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,train_size = 0.70,random_state = 22)
```

```python
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
model = LinearRegression()
model.fit(x_train,y_train)
```
