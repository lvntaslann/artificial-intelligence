```python
for i in range(len(df)):
    df.at[i, 'Height'] = df.at[i, 'Height'] * 100

df.Height
```

```python
for i in range(len(df)):
    df.at[i, 'Weight'] = int(df.at[i, 'Weight'])

df.Weight
```

```python
df.head()
```

```python
df.info()
```

```python
from sklearn.preprocessing import LabelEncoder
gender_label = LabelEncoder()
calc_label = LabelEncoder()
favc_label = LabelEncoder()
scc_label = LabelEncoder()
smoke_label = LabelEncoder()
family_history_with_overweight_label = LabelEncoder()
caec_label = LabelEncoder()
mtrans_label = LabelEncoder()
Nobeyesdad_label = LabelEncoder()

df['Gender'] = gender_label.fit_transform(df['Gender'])
df['CALC'] = calc_label.fit_transform(df['CALC'])
df['FAVC'] = favc_label.fit_transform(df['FAVC'])
df['SCC'] = scc_label.fit_transform(df['SCC'])
df['SMOKE'] = smoke_label.fit_transform(df['SMOKE'])
df['family_history_with_overweight'] = family_history_with_overweight_label.fit_transform(df['family_history_with_overweight'])
df['CAEC'] = caec_label.fit_transform(df['CAEC'])
df['MTRANS'] = mtrans_label.fit_transform(df['MTRANS'])
df['NObeyesdad'] = Nobeyesdad_label.fit_transform(df['NObeyesdad'])

df.head()
```

```python
from sklearn.model_selection import train_test_split
x = df.drop('NObeyesdad',axis=1)
y= df['NObeyesdad']
x = x.astype('float64')

x_train,x_test,y_train,y_test = train_test_split(x,y,train_size = 0.70,random_state=42)
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression

model1 = RandomForestClassifier()
model2 = LogisticRegression()
```

```python
model1.fit(x_train,y_train)
```

```python
model2.fit(x_train,y_train)
```

```python
model1.score(x_test,y_test)
```

```python
model2.score(x_test,y_test)
```

```python
tahmin1 = model1.predict(x_test.iloc[[0]])
tahmin2 = model2.predict(x_test.iloc[[0]])
print(tahmin1)
print(tahmin2)
```
