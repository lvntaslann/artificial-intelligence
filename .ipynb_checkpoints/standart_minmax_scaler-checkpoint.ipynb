```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler,MinMaxScaler
```

```python
df = pd.read_csv('plane_price.csv')
```

```python
df = df.dropna()
```

```python
df.info()
```

```python
df.head(3)
```

```python
engine_type = df['Engine Type'].unique()
print(engine_type)
```

```python
df['Engine Type'] = df['Engine Type'].replace({'Piston':0,'Propjet':1,'Jet':2,'piston':3})
df.head(3)
```

```python
df=df[["Rcmnd cruise Knots","Engine Type","Stall Knots dirty","Fuel gal/lbs","Eng out rate of climb","Price"]]
```

```python
df.head()
```

```python
y = df['Price']
x = df.drop("Price",axis = 1)
```

```python
standard_scaler = StandardScaler()
x_standardScaler = standard_scaler.fit_transform(x)
x_standardScaler = pd.DataFrame(x_standardScaler)
x_standardScaler.head()
```

```python
# ortalama sıfıra yaklaşırken standart sapmayı 1 e sabitliyor
print(x_standardScaler[2].mean())
print(x_standardScaler[2].std())
```

```python
minMaxScaler = MinMaxScaler()
x_minMaxScaler = minMaxScaler.fit_transform(x)
x_minMaxScaler = pd.DataFrame(x_minMaxScaler)
x_minMaxScaler.head()
```

```python
print(x_minMaxScaler[2].min())
print(x_minMaxScaler[2].max())
```
